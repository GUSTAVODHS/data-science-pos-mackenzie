 **Introdução**
        
Nessa tarefa utilizei um texto narrativo sobre a história da terra.

É válido ressaltar que, as maiores ocorrências que podemos prever nesse texto são o emprego das palavras 'a' e 'de'. Isso acontece pois se trata da história da **Terra** que é um substantivo feminino.

```python
texto = []

f = open('/historia_terra.txt','r',encoding='iso-8859-1')    # para arquivos locais
for line in f:
    texto.append(line)

for i in range(len(texto)):
    texto[i] = texto[i].lower() # para unicidade
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('\'','') # elimina ' e "
    
```

```python
dicionario = {}                             # crie um dicionário vazio

for line in texto:

  line = line.lower()                   # converte para lower
  words = line.split()                  # separa cada palavra
#  print(words)

  for word in words:                      # para cada palavra em words
    if word not in dicionario.keys():         # se palavra não está no dicionário 
      dicionario[word] = 1                    # acrescenta a word com o valor 1
    else:                                 # se a entrada já existe 
      dicionario[word] =  dicionario[word] + 1    # apenas soma 1 ao valor já existente

print(dicionario)    
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(dicionario.items(),  columns=['word', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 4] # somente termos com mais de 4 ocorrências
# df = df.iloc[ np.int(len(df)/2) - 10 : np.int(len(df)/2) + 10 ] # para livros ou textos com muitos termos limita a um número mínimo de termos

plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word,y=df['count'])
plt.xticks(rotation=90) 

plt.show()
```

**Conclusão**
        
A partir da Lei de Zipf é possível notar que existe um padrão matemático na frequência de uso das palavras. Nesse caso estou utilizando um texto em português. Porém, esse padrão se repete nas demais línguas.

Na minha opinião, acho que isso ressalta o fato de que independente das diferenças culturais dos seres humanos, existem laços fortes que unem todas as línguas.
